```python
# Dependencies
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
# Read CSV

kings_data_load = "Resources/got.csv"
kings_df= pd.read_csv(kings_data_load)
kings_df
```

```python
# Get attacker and defender king data
attacker_king= kings_df["attacker_king"].value_counts() 

defender_king= kings_df["defender_king"].value_counts()
```

```python
# Get total battle data
Total_data= attacker_king.add(defender_king,fill_value=0)
Total_data
```

```python
# Configure plot and ticks
Total_data.plot(kind="bar", figsize=(8, 6), facecolor="red")

# Set textual properties
plt.xlabel('Kings')
plt.ylabel('Number of Battle Partcipated in')
plt.title('The Bloddthirst of Kings')
plt.xticks(rotation=45)
# Show plot
plt.show()
plt.legend()
# Resize plot to display labels
plt.tight_layout()
```
